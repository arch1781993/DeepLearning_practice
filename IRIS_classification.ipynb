```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import pandas as pd
from sklearn.model_selection import train_test_split
```

# Load Dataset

```python
url = 'https://gist.githubusercontent.com/curran/a08a1080b88344b0c8a7/raw/0e7a9b0a5d22642a06d3d5b9bcbad9890c8ee534/iris.csv'
df = pd.read_csv(url)
```

```python
train_df, test_df = train_test_split(df,test_size = 0.2, random_state =42)
```

```python
replace = {'setosa':0, 'versicolor':1, 'virginica':2}
train_df['species'] = train_df['species'].map(replace)
test_df['species'] = test_df['species'].map(replace)
```

```python
input_cols = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
target_cols = 'species'
train_inputs = train_df[input_cols]
train_targets = train_df[target_cols]
test_inputs = test_df[input_cols]
test_targets = test_df[target_cols]
```

```python
# Transforming to tensors
# First transform to Numpy arrays
train_inputs = train_inputs.values
train_targets = train_targets.values
test_inputs = test_inputs.values
test_targets = test_targets.values
```

```python
# Now transform to tensors
train_inputs = torch.FloatTensor(train_inputs)
train_targets = torch.LongTensor(train_targets)
test_inputs = torch.FloatTensor(test_inputs)
test_targets = torch.LongTensor(test_targets)
```

# Model Architecture

```python
class Model(nn.Module):

  def __init__(self, in_features =4 , h1 = 16 , h2 = 8 , out_features = 3):
    super().__init__()
    self.fc1 = nn.Linear(in_features,h1)
    self.fc2 = nn.Linear(h1,h2)
    self.out = nn.Linear(h2,out_features)

  def forward(self,X):
    X = F.relu(self.fc1(X))
    X = F.relu(self.fc2(X))
    X = self.out(X)
    return X
```

```python
# Pick a manual seed for randomization
torch.manual_seed(42)
model = Model()
```

```python
print(model)
```

```python
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.01)
```

```python
epochs = 100
for i in range(epochs):
  train_preds = model(train_inputs)
  loss = criterion(train_preds,train_targets)

  if i%10==0:
    print(f'Epoch: {i}  Loss: {loss}')

  optimizer.zero_grad()
  loss.backward()
  optimizer.step()
```
